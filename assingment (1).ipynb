```python
#Q4. What is a generator function in python? Why yield keyword is used? Give an example of a generator
#function.
#ans4 A generator function in Python is a type of function that uses the yield keyword to return an iterator object. The yield keyword allows the generator function to return a value and then resume execution from where it left off when the function is called again. 
#This makes it possible to iterate over a sequence of values generated by the function without having to store the entire sequence in memory.
```

```python
def test_fib(n):
    a,b=0,1
    for i in range(n):
        yield a
        a,b=b , a+b
    
```

```python
test_fib(10)
```

```python
for i in test_fib(10):
    print(i)
```

```python
#Q5. Create a generator function for prime numbers less than 1000. Use the next() method to print the
#first 20 prime numbers.
```

```python
def primes():
    for num in range(2, 1000):
        if all(num % i != 0 for i in range(2, int(num ** 0.5) + 1)):
            yield num
```

```python
#Q8. Write a python program to check whether a given number is Palindrome or not using a while loop.
```

```python
num=int(input("Enter any number:"))
temp=num
rev=0
while(num>0):
    dig=num%10
    rev=rev*10+dig
    num=num//10
if(temp==rev):
    print("The {0} number is palindrome!".format(temp))
else:
    print("Not a palindrome!")
```
